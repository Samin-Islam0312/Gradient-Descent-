```python
import numpy as np

x = np.random.randn(10,1)
y = 2*x + np.random.randn()
#parameters
w = 0.0
b =0.0
#Hyperparameter
lr = 0.01
```

```python
# Create function
def descend(x, y, w, b, lr):
  dl_dw = 0.0 #derivative in respect to weight parameter
  dl_db = 0.0 #derivative in respeect to bias parameter
  N = x.shape[0]

  for i,j in zip(x,y):
    dl_dw += -2*i*(j-(w*i+b))
    dl_db += -2*(j-(w*i+b))

  w = w-lr*(1/N)*dl_dw
  b = b-lr*(1/N)*dl_db
  return w,b 
```

```python
#iteratively make updates
for epoch in range(400):
  #run gradient descent
  w,b = descend(x,y,w,b,lr)
  y_hat = w*x+b
  loss = np.divide(np.sum((y-y_hat)**2, axis=0), x.shape[0])
  print(f'{epoch} loss is {loss}, parameters w: {w}, b: {b}')
```
